# (키보드를 통해) 스크롤 하는 법
- 스크롤을 페이지 최하단으로 이동
- 키보드에 있는 end키를 보내기!

```python
# body태그를 먼저 알아오기 !
body = driver.find_element(By.TAG_NAME,"body")
body.send_keys(Keys.END)    # 키값 "" 하는거 아님 ㅜㅜ
```

```python
# 한번에 정리하기
driver = wb.Chrome()
time.sleep(0.5)
driver.get("https://www.youtube.com/c/%EB%B9%A0%EB%8B%88%EB%B3%B4%ED%8B%80PaniBottle/videos")
time.sleep(1)

body = driver.find_element(By.CSS_SELECTOR,"body")
for i in range(10):
    body.send_keys(Keys.END)
    time.sleep(1)
    
title = driver.find_elements(By.ID, "video-title")
view = driver.find_elements(By.CSS_SELECTOR, "span.style-scope.ytd-grid-video-renderer")
titleList = []
viewList = []
for i in range(len(title)):
    titleList.append(title[i].text)
    viewList.append(view[i].text)
print(titleList)

driver.quit()
```
