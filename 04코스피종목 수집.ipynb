```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
# 1. 네이버 금융사이트에서 코스피 TOP100사이트의 정보를 요청
# 익숙해졌다고 바로 변수에 할당X ->>  <Response [200]> 확인 후 변수 할당!!
res = req.get("https://finance.naver.com/sise/sise_quant.naver")
```

```python
res.text
```

```python
# 2. 사용자가 받아온 데이터(text) 컴퓨터가 알아 먹을 수 있게 컴퓨터용 언어로 변형!
# html언어로 변형 > bs라이브러리!
soup = bs(res.text,"lxml")
```

```python
soup
```

```python
# 3. soup 데이터에서 내가 필요한 요소를 찾아오기!!
title = soup.select("a.tltle")
```

```python
price = soup.select("td.number")
```

```python
price 
```

```python
# 4. 데이터의 개수가 정확한지 꼭 확인 !
# 리스트의 데이터 길이를 확인! >> len()
len(title)
# 잘 가져옴
```

```python
len(price)
# 잘못 가져옴
```

```python
# 가상 클래스 선택자
price = soup.select("td.number:nth-child(1)")
```

```python
len(price)
# 잘못 가져옴
```

```python
for i in title:
    print(i.text)
```

```python
# title, price, volume 리스트에서 순수한 글자 정보만 추출한 리스트를 새로 제작
# range(len(title)) > 좀 더 동적으로 반복문 제어 가능 

titleList =[]
priceList =[]
volumeList =[]

for i in range(len(title)):
    titleList.append(title[i].text)
    priceList.append(price[i].text)
    volumeList.append(volume[i].text)
```

```python
titleList
```

```python
len(titleList)
```

```python
len(priceList)
```

```python
len(volumeList)
```

# 정제된( 순수한 글자 ) 데이터를 표 형식 제작

```python
import pandas as pd
```

```python
data = {"종목명": titleList, "가격정보": priceList, "거래량": volumeList}
```

```python
kospi = pd.DataFrame(data)
```

```python
# 한글 사용을 위해선 encoding까지 거쳐야 함
# 1. euc-kr
# 2. ""
# 3. utf-8-sig
kospi.to_csv("코스피TOP100.csv", encoding = "euc-kr")
```
