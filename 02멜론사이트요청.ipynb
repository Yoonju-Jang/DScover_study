```python
import requests as req
```

```python
# <Response [406]> = 요청은 했지만 통신에 실패했습니다!
req.get("https://www.melon.com/")
```

```python
# headers 작업 : 서버쪽에게 브라우저로 접근한척 속이는 작업
# User-Agent의 유무로 브라우저로 접근했는지 판단
# F12 > Network + F5 > 맨 윗 상단에 있는 docx 클릭 >  맨
head = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36'}
```

```python
# 크롤링하다 200번이 아닌 400번대 응답이 나오면 -> headers 작업을 추가
res = req.get("https://www.melon.com/", headers = head)
```

```python
res.text
# requests로 받은 데이터는 text데이터(사람을 위한 정보)
```
