```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
res = req.get("https://movie.naver.com/movie/bi/mi/point.naver?code=208077") # 잘못된 사이트 요청
```

```python
res = req.get("https://movie.naver.com/movie/bi/mi/pointWriteFormList.naver?code=208077&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false")
```

```python
soup= bs(res.text,"lxml")
```

```python
# 영화 관람객들의 리뷰데이터를 수집
```

```python
# id가 아닌 class로 접근하는게 가장 좋은 방법이다. 
# id는 고유값으로 데이터를 1개만 수집 가능
soup.select("div.score_reple> p")
```

```python
# 내가 원하는 요소들이 아니다! 
# 내가 원하는 요소가 실제로는 다른 서버에 저장되어 있는 것이다 ! <iframe>
# Ctrl + F > iframe > src 주소 복붙 > 주소 변경
```

```python
review = soup.select("div.score_reple> p")
```

```python
# 불필요한 엔터 삭제하는 방법 : strip()
for i in review :
    print(i.text.strip())
```

```python
viewer = soup.select("span.ico_viewer")
```

```python
# 불필요한 데이터(요소) 추출해서 제거하는 메소드 : extract()
for i in viewer:
    i.extract()  # soup에 자동으로 영향
```

```python
for i in review :
    print(i.text.strip())
```

```python
# class="text_spo _text_spo" -> 클래스 2개 보유중 -> 가운데 enter를 지우고 .으로 이어주기
spo = soup.select("span.text_spo._text_spo")
```

```python
for i in spo:
    i.extract()
```

```python
for i in review :
    print(i.text.strip())
```

```python
# 1~10페이지 추출
# 하나의 완벽한 로직을 완성시킨 뒤 -> for문 생성 !!
```

```python
for i in range(1,11):
    
    res = req.get("https://movie.naver.com/movie/bi/mi/pointWriteFormList.naver?code=208077&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page="+str(i))
    soup= bs(res.text,"lxml")
    viewer = soup.select("span.ico_viewer")
    spo = soup.select("span.text_spo._text_spo")
    for j in viewer:
        j.extract() 
    for p in spo:
        p.extract()
    review = soup.select("div.score_reple> p")
    for i in review :
        print(i.text.strip())
```

## 번외) 반복문의 진행상황을 체크하는 라이브러리
- 반복의 진행상황을 로딩바 형태로 확인하는 라이브러리
- tqdm 라이브러리

```python
! pip install tqdm
```

```python
from tqdm import tqdm_notebook as tq
```

```python
for i in tq(range(1,30)):
    
    res = req.get("https://movie.naver.com/movie/bi/mi/pointWriteFormList.naver?code=208077&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page="+str(i))
    soup= bs(res.text,"lxml")
    viewer = soup.select("span.ico_viewer")
    spo = soup.select("span.text_spo._text_spo")
    for j in viewer:
        j.extract() 
    for p in spo:
        p.extract()
    review = soup.select("div.score_reple> p")
    for i in review :
        print(i.text.strip())
```
